# The Marriage Problem
This was inspired by Lecture 8: Fixed Income and LTCM of BEM 114: Hedge Funds by Prof. Andrew Sinclair.

## Compatibility Generation

```python
import matplotlib.pyplot as plt

N = 1000 # number of people total
Bs = [0] * (N+1) # list of min acceptable
Bs[0] = 0.5 # the initial B
for i in range(N): # loop N times
  Bs[i+1] = 1/2 + pow(Bs[i], 2)/2 # the formula described above
```

```python
sample_Ns = [1, 10, 50, 100, 500, 1000]
for n in sample_Ns:
    print(f"Minimum acceptible compatibilities at {n} people remaining: {Bs[n-1]:.3f}")
```

## Visualization

```python
plt.plot(Bs, label='Optimal Standards')
plt.plot([0.7] * 1000, label='People\'s Standards')
plt.legend()
plt.title("Marriage Problem Min Acceptable Compatibility")
plt.ylabel("Minimum Acceptable Compatibility")
plt.xlabel("Number of People Remaining (N)")
plt.savefig("marriage_line_plot.png")
```
